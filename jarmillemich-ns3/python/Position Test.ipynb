```python
pts = []
ptInterval = MilliSeconds(int(10000))

def logPoint():
    pt = mobile.GetPosition()
    pts.append((pt.x, pt.y, pt.z))
    ns.core.Simulator.Schedule(ptInterval, logPoint)
    
from tqdm.notebook import tqdm
pbar = tqdm(unit='ms')
    
def logTime():
    now = ns.core.Simulator.Now().GetSeconds()
    #print('T=%d' % (now))
    pbar.n = int(now)
    pbar.update(int(0))
    ns.core.Simulator.Schedule(MilliSeconds(int(500)), logTime)
    
pbar.reset(total = int(6000))

ns.core.Simulator.Destroy()
ns.core.Simulator.Schedule(Seconds(mobile.getDuration()), logPoint)
#ns.core.Simulator.Schedule(Seconds(2256), logPoint)
logTime()

ns.core.Simulator.Stop(Seconds(mobile.getDuration() * 8))
#ns.core.Simulator.Stop(Seconds(int(2260)))
ns.core.Simulator.Run()
pbar.refresh()

#show(pts[100])
pts = np.array(pts) / 1e3
list_plot(pts) + point((0,0,0), color='red', size=50)
```
